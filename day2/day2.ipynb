```python
file1 = open('input.txt', 'r')
lines = file1.readlines()
```

```python
def compute_score(player1,player2):
  
  if player2 == 'X':

      score = 1
      if player1 == 'A':
        score = score + 3
      if player1 == 'B':
        score = score + 0
      if player1 == 'C':
        score = score + 6

  elif player2 == 'Y':

      score = 2

      if player1 == 'A':
        score = score + 6
      if player1 == 'B':
        score = score + 3
      if player1 == 'C':
        score = score + 0

  elif player2 == 'Z':

      score = 3

      if player1 == 'A':
        score = score + 0
      if player1 == 'B':
        score = score + 6
      if player1 == 'C':
        score = score + 3



  return score

```

```python
total_score = 0

for line in lines:
  content = line.strip().split(" ")
  #print(content[0]," ",content[1]," ",compute_score(content[0],content[1]))
  total_score = total_score + compute_score(content[0],content[1])

print("final score : ",total_score)
```

```python
def choose_move(player1,player2):

    if player2 == 'X': # lose

      
      if player1 == 'A':
        return 'Z'
      if player1 == 'B':
        return 'X'
      if player1 == 'C':
        return 'Y'

    elif player2 == 'Y': # draw

      
      if player1 == 'A':
        return 'X'
      if player1 == 'B':
        return 'Y'
      if player1 == 'C':
        return 'Z'

    elif player2 == 'Z': # win      

      if player1 == 'A':
        return 'Y'
      if player1 == 'B':
        return 'Z'
      if player1 == 'C':
        return 'X'
```

```python
total_score = 0

for line in lines:
  
  content = line.strip().split(" ")  
  move = choose_move(content[0],content[1])
  total_score = total_score + compute_score(content[0],move)

print("final score : ",total_score)
```
